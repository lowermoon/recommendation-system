```python
import pandas as pd
from faker import Faker
import random
```

```python
#create an object faker
fake = Faker()
```

```python
#steam_review dataset for project_desc entries
review_data = pd.read_csv('/content/drive/My Drive/ml-syst-recommend/review.csv')
```

```python
# preference genre
genres = ['RPG', 'FPS', 'MOBA', 'Strategi', 'Aksi/Petualangan', 'Battle Royal']

#generate freelancer_data
freelancer_data = []
for i in range(1, 3001):
    preferences = ', '.join(random.sample(genres, random.randint(1, 3)))
    freelancer_data.append({'freelancer_id': f"freelancer_{i}", 'preference': preferences})

freelancer_df = pd.DataFrame(freelancer_data)
```

```python
print(freelancer_df.head())
```

```python
game_descriptions = [
    "Looking for a game booster {} to reach the highest level",
    "Join the {} team and help us win the tournament!",
    "We need help mastering {} for a big victory!",
    "The excitement of {} is waiting for you, come join us!",
    "Be the best in {} and achieve the highest accomplishments!",
    "Focus on pro player game {}",
    "GG player {} over here",
    "Victory with us for {}"
]
#generate active_project_data
active_project_data = {
    'nama_project': [random.choice(["joki {}", "cari {}", "tutor {}"]).format(fake.word()) for _ in range(2000)],
    'desc_project': [random.choice(game_descriptions).format(fake.word()) for _ in range(2000)],
    'offer_desc': [fake.name() for _ in range(2000)],
    'offer_price': [random.randint(10000, 500000) for _ in range(2000)]
}
active_project_df = pd.DataFrame(active_project_data)
```

```python
random_2000_entries = review_data.sample(n=2000)
```

```python
# selected 2000 entries from review_data to subset in reviews_subset
reviews_subset = random_2000_entries['review']
```

```python
%store project_df
```

```python
categories = ['valorant', 'mobile legend', 'free fire', 'cewek', 'jodoh', 'cogan', 'cecan', 'waifu', 'wibu', 'gg', 'pro player', 'susah', 'sepuh', 'suhu']
games = ['ML', 'FF', 'VALORANT', 'NINTENDO', 'Minecraft', 'PUBG', 'Roblox', 'Bang Bang', 'Genshin Impact', 'Clash Royale', 'DOTA', 'GTA', 'Resident Evil', 'COD', 'Cyberpunk', 'Destiny']

#generate project_data
project_data = {
    'name_project': [random.choice(["joki {}", "cari {}", "tutor {}"]).format(random.choice(games)) for _ in range(2000)],
    'project_desc': reviews_subset,
    'category_project': [', '.join(random.sample(categories, random.randint(1, 3))) for _ in range(2000)]
}
project_df = pd.DataFrame(project_data)
```

```python
print(project_df.head())
```

```python
random_2000_entries = review_data.sample(n=2000)
```

```python
# selected 2000 column from review_data to subset in reviews_subset_for_offer
reviews_subset_for_offer = random_2000_entries['review']
```

```python
#generate offer_project_data
offer_project_data = {
    'project_name': active_project_df['nama_project'],
    'project_desc': reviews_subset_for_offer,
    'offer_desc': active_project_df['offer_desc']
}
offer_project_df = pd.DataFrame(offer_project_data)
```

```python
print(offer_project_df.head())
```

```python
freelancer_df.to_csv('/content/freelancer_data.csv', index=False)
active_project_df.to_csv('/content/active_project_data.csv', index=False)
project_df.to_csv('/content/project_data.csv', index=False)
offer_project_df.to_csv('/content/offer_project_data.csv', index=False)
```
