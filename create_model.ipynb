```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.externals import joblib
```

```python
project_df = pd.read_csv('/content/drive/My Drive/ml-syst-recommend/data/project_data.csv')
```

```python
# Deskripsi proyek
project_descriptions = project_df['project_desc'].tolist()
project_names = project_df['name_project'].tolist()
```

```python
# Menggabungkan deskripsi proyek dan nama proyek
combined_descriptions = [f"{desc} {name}" for desc, name in zip(project_descriptions, project_names)]

# Inisialisasi TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words='english')

# Vektorisasi deskripsi proyek
tfidf_matrix = tfidf_vectorizer.fit_transform(combined_descriptions)
```

```python
# Menghitung kesamaan kosinus antar proyek
similarities = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
# Contoh rekomendasi untuk proyek dengan index tertentu (misalnya, index 0)
project_index = 0
similar_projects = similarities[project_index].argsort()[-5:-1][::-1]  # Ambil 4 proyek teratas (kecuali dirinya sendiri)

for index in similar_projects:
    print(f"Nama Proyek: {project_df.iloc[index]['name_project']}")
    print(f"Deskripsi: {project_df.iloc[index]['project_desc']}")
    print("-----")
```

```python
# Ambil data yang akan digunakan untuk pengujian
test_project_index = 10  # Misalnya, mengambil salah satu proyek sebagai "data uji"
```

```python
# Ambil data proyek uji dari tfidf_matrix
test_project_data = tfidf_matrix[test_project_index]
test_project_data = test_project_data.reshape(1, -1)  # Reshape data ke bentuk yang sesuai
```

```python
# Hitung kesamaan kosinus antara proyek uji dan proyek lainnya
similarities_to_test = cosine_similarity(test_project_data, tfidf_matrix)
```

```python
# Ambil beberapa proyek teratas (kecuali dirinya sendiri) sebagai rekomendasi
top_similar_projects = similarities_to_test.argsort()[0][-6:-1][::-1]
```

```python
# Tampilkan rekomendasi
for index in top_similar_projects:
    print(f"Nama Proyek: {project_names[index]}")
    print(f"Deskripsi: {project_descriptions[index]}")
    print("-----")
```

```python
# Hitung kesamaan kosinus antara proyek pengujian dan semua proyek lainnya
similarities_to_test = cosine_similarity(tfidf_matrix[test_project_index], tfidf_matrix)
```

```python
# Indeks proyek-proyek yang memiliki kesamaan kosinus tertinggi dengan proyek pengujian
similar_projects_indices = similarities_to_test.argsort()[0][-5:-1][::-1]
```

```python
# Ambil label atau kategori dari proyek-proyek yang paling mirip
similar_projects_categories = project_df.iloc[similar_projects_indices]['category_project']
```
